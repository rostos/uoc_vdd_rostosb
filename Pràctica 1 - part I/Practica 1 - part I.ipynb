```python
import csv

hpo_data = []

with open("Habitatges_amb_protecci__oficial_iniciats_i_acabats._S_rie_hist_rica_anual_2002_-_actualitat_20240502.csv", 'r', encoding='utf-8-sig') as f:
    atributs = f.readline().replace('\n','').split(',')
    # Quan llegim els valors, voldrem llegir-los de forma numèrica i a més ignorar els dos primers com hem fet al pas anterior
    for line in csv.reader(f):
        hpo_data.append(line)
    
hpo_data
```

```python
atributs
```

```python
hpo_data_2 = []

for line in hpo_data:
    if line[9] != '0' or line[10] != '0':
        hpo_data_2.append(line)

len(hpo_data_2)
```

```python
writer = csv.writer(open("hpo_data_clean.csv", 'w', encoding='utf-8-sig', newline=''))

writer.writerow(atributs)

for row in hpo_data_2:
    writer.writerow(row)     
```
